```typescript
function bookSearchPromise(){
    return new Promise((resolve,reject)=>{
        setTimeout(()=>{
            let bookFound=Math.random()>0.5
            if(bookFound){
                resolve("Book found")
            }
            else {
                reject("book not found")
            }
        },3000)
    })
}
bookSearchPromise()
.then((res)=>console.log(res))
.catch((err)=>console.log(err))
```

```typescript
let closestToArr=function(arr){
    return arr.sort((a,b)=>{
        let A=Math.abs(a-5)
        let B=Math.abs(b-5)
        return A-B
    })
}
let arr=[1,3,5,7,9]
console.log(closestToArr(arr))
```

```typescript
let countVowel=function(str){
    str=str.toLowerCase()
    let count=0
    let vowel=['a','e','i','o','u']
    for(let i=0;i<str.length;i++){
        if(vowel.includes(str[i])){
            count++
        }
    }
    return count
}
let str="a1e@i2o!u"
console.log(countVowel(str))
```

```typescript
function findDuplicate(arr){
    arr=arr.sort((a,b)=>a-b)
    for(let i=0;i<arr.length;i++){
        if(arr[i] === arr[i+1])return "True"
    }
    return "False"
}
let arr=[1,2,3,4]
console.log(findDuplicate(arr))
```

```typescript
function findDuplicate(arr){
    let set =new Set()
    for(let i=0;i<arr.length;i++){
        if(set.has(arr[i]))return "True"
        else set.add(arr[i])
    }
    return "False"
}
let arr=[1,2,3,1]
console.log(findDuplicate(arr))
```

```typescript
function divisibleBy69(num){
    if(num === 69)return 0
    let div=num%69
    let ans=69-div
    return ans
}
let num=123
console.log(divisibleBy69(num))
```

```typescript
function rotatoria(arr){
    let left=0,right=arr.length-1
    while(left < right){
        let mid=Math.floor((left+right)/2)
        if(arr[mid] < arr[right]){
            right=mid
        }
        else{
            left=mid+1
        }
    }
    return arr[left]
}
let arr=[4,5,6,7,0,2]
console.log(rotatoria(arr))
```

```typescript
let showTime=function(n,k,movieTime){
    movieTime.sort((a,b)=>a[1]-b[1])
    let recessTime=new Array(k).fill(0),count=0
    for(let i=0;i<n;i++){
        let [start,end]=movieTime[i]
        let index=-1
        for(let j=0;j<k;j++){
            if(recessTime[j] < start){
                index=j
                break
            }
        }
        if(index !== -1){
            recessTime[index]=end
            count++
        }
    }
    return count
}
let n=5,k=2,movieTime=[[1,5],[8,10],[3,6],[2,5],[6,9]]
console.log(showTime(n,k,movieTime))
```
