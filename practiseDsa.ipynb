```typescript
let sortIntByOnesBit=function(arr){
    return arr.sort((a,b)=>{
        let A=a.toString(2).split('').filter(i=>i==='1').length
        let B=b.toString(2).split('').filter(i=>i==='1').length
        return A-B || a-b 
    })
}
let arr = [0,1,2,3,4,5,6,7,8]
console.log(sortIntByOnesBit(arr))
```

```typescript
let divideTwoInteger=function(dividend,divisor){
    let res=parseInt(dividend/divisor)
    if(res > 2147483647){
        return 2147483647
    }
    else if(res < -2147483648){
        return -2147483648
    }
    else return res
}
let d=10,di=3
console.log(divideTwoInteger(d,di))
```

```typescript
let binarySearch=function(nums,target){
    let left=0,right=nums.length-1
    while(left <= right){
        let mid=Math.floor((left+right)/2)
        if(nums[mid] === target){
            return mid
        }
        else if(nums[mid] > target){
            right=mid-1
        }
        else{
            left=mid+1
        }
    }
    return -1
}
let  nums = [-1,0,3,5,9,12], target = 9
console.log(binarySearch(nums,target))
```

```typescript
function search(nums,target){
    let left=0,right=nums.length-1
    while(left <= right){
        let mid=Math.floor((left+right)/2)
        if(nums[mid] === target)return mid
        else if(nums[mid] > nums[left]){
            if(target <= nums[mid] && target >= nums[left]){
                right=mid-1
            }
            else{
                left=mid+1
            }
        }
        else{
            if(target >= nums[mid] && target <= nums[right]){
                left=mid+1 
            }
            else{
                right=mid-1 
            }
        }
    }
    return -1
}
let nums = [4,5,6,7,0,1,2], target = 0
console.log(search(nums,target))
```

```typescript
let twoSum=function(nums,target){
    for(let i=0;i<nums.length;i++){
        for(let j=i+1;j<nums.length;j++){
            if((nums[i]+nums[j])===target){
                return [i,j]
            }
        }
    }
    return [-1,-1]
}
let  nums = [2,7,11,15], target = 9
console.log(twoSum(nums,target))
```

```typescript
class Employee{
    constructor(name,salary){
        this.name=name
        this.salary=salary
    }
}
let emp=[
    new Employee('Babu_Rao',150),
    new Employee("Raju",160),
    new Employee("Shyam",140),
    new Employee("Pappu",200)
]
function sortEmployee(emp){
    for(let i=0;i<emp.length;i++){
        let temp=emp[i]
        let j=i-1 
        while(j>=0 && emp[j].salary > temp.salary){
            emp[j+1]=emp[j]
            j--
        }
        emp[j+1]=temp
    }
    return emp
}
console.log(sortEmployee(emp))
```

```typescript
let countPrime=function(n){
    let isPrime=new Array(n).fill(true)
    isPrime[0]=isPrime[1]=false
    for(let i=2;i*i<n;i++){
        if(isPrime[i]){
            for(let j=i*i;j<n;j+=i){
                isPrime[j]=false
            }
        }
    }
    return isPrime.filter(prime=>prime).length
}
let n=10
console.log(countPrime(n))
```

```typescript
let validParentheses=function(s){
    let dict={']':'[','}':'{',')':'('},stack=[]
    for(let i of s){
        if(i in dict){
            if(stack.length === 0 || stack.pop() !== dict[i])return false
        }
        else stack.push(i)
    }
    return stack.length === 0
}
let s="{{}}"
console.log(validParentheses(s))
```

```typescript
let validateStackSeq=function(pushed,popped){
    let stack=[]
    for(let i  of pushed){
        stack.push(i)
        while(stack.length && stack[stack.length-1] === popped[0]){
            stack.pop()
            popped.shift()
        }
    }
    return stack.length === 0
}
let pushed = [1,2,3,4,5], popped = [4,5,3,2,1]
console.log(validateStackSeq(pushed,popped))
```

```typescript
let removeAdjacent=function(s){
    let stack=[]
    for(let i of s){
        if(stack.length && stack[stack.length-1]===i){
            stack.pop()
        }
        else{
            stack.push(i)
        }
    }
    return stack.join('')
}
let s="abbaca"
console.log(removeAdjacent(s))
```

```typescript
let decodeString=function(s){
    let stack=[],num=0,str=''
    for(let i of s){
        if(!isNaN(i)){
            num=num*10+parseInt(i)
        }
        else if(i === '['){
            stack.push(str)
            stack.push(num)
            str=''
            num=0
        }
        else if(i === ']'){
            let prevNum=stack.pop()
            let prevStr=stack.pop()
            str=prevStr+str.repeat(prevNum)
        }
        else{
            str+=i 
        }
    }
    return str
}
let s="3[a]2[bc]"
console.log(decodeString(s))
```

```typescript
let nextGreaterElement=function(nums){
    let arr=[]
    for(let i=0;i<nums.length;i++){
        arr[i]=-1
        for(let j=1;j<nums.length;j++){
            let index=(i+j)%nums.length
            if(nums[index]> nums[i]){
                arr[i]=nums[index]
                break
            }
        }
    }
    return arr
}
let nums=[1,2,1]
console.log(nextGreaterElement(nums))
```

```typescript
let stockSpan=function(){
    this.stack=[]
}
stockSpan.prototype.next=function(price){
    let span=1
    while(this.stack.length && this.stack[this.stack.length-1].price <= price){
        span+=this.stack.pop().span
    }
    this.stack.push({price,span})
    return span
}
```

```typescript
let largestRectangleHisto=function(heights){
    let stack=[-1],maxi=0
    for(let i=0;i<heights.length;i++){
        while(heights[i] < heights[stack[stack.length-1]]){
            let left=heights[stack.pop()]
            let right=i-stack[stack.length-1]-1
            maxi=Math.max(maxi,(left*right))
        }
        stack.push(i)
    }
    return maxi
}
let heights=[2,1,5,6,2,3]
console.log(largestRectangleHisto(heights))
```

```typescript
class Queue{
    constructor(){
        this.queue=[]
    }
    Enqueue(val){
        this.queue.push(val)
    }
    Dequeue(){
        if(!this.is_empty()){
            return this.queue.shift()
        }
    }
    is_empty(){
        return this.queue.length === 0
    }
    get_first_element(){
        if(!this.is_empty()){
            return this.queue[0]
        }
    }
}
let q=new Queue()
q.Enqueue(12)
q.Enqueue(24)
q.Enqueue(36)
q.Dequeue()
q.is_empty()
q.get_first_element()
```

```typescript
let nonRepeatingChar=function(A){
    let freq=new Array(26).fill(0)
    let str='',stack=[]
    for(let i of A){
        let index=i.charCodeAt(0)-'a'.charCodeAt(0)
        freq[index]++
        stack.push(i)
        while(stack.length && freq[stack[0].charCodeAt(0)-'a'.charCodeAt(0)]>1){
            stack.shift()
        }
        if(stack.length){
            str+=stack[0]
        }
        else{
            str+='#'
        }
    }
    return str
}
let A="aabc"
console.log(nonRepeatingChar(A))
```

```typescript
let slidingWIndowMaximum=function(nums,k){
    let queue=[],ans=[]
    for(let i=0;i<nums.length;i++){
        while(queue.length && queue[0] < i-k+1){
            queue.shift()
        }
        while(queue.length && nums[queue[queue.length-1]]<= nums[i]){
            queue.pop()
        }
        queue.push(i)
        if(i>=k-1){
            ans.push(nums[queue[0]])
        }
    }
    return ans
}
let nums = [1,3,-1,-3,5,3,6,7], k = 3
console.log(slidingWIndowMaximum(nums,k))
```
