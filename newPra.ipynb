```typescript
function rotatedArray(nums,target){
    let left=0,right=nums.length-1
    while(left <= right){
        let mid=Math.floor((left+right)/2)
        if(nums[mid] === target)return mid
        else if(nums[mid] > nums[left]){
            if(target <= nums[mid] && target >= nums[left]){
                right=mid-1
            }
            else{
                left=mid+1
            }
        }
        else{
            if(target >= nums[mid] && target <= nums[right]){
                left=mid+1
            }
            else{
                right=mid-1
            }
        }
    }
    return -1
}
let nums=[4,5,6,7,0,1,2], target = 0
console.log(rotatedArray(nums,target))
```

```typescript
let mergeSort=function(arr){
    if(arr.length <= 1)return arr
    let mid=Math.floor(arr.length/2)
    let left=mergeSort(arr.slice(0,mid))
    let right=mergeSort(arr.slice(mid))
    return merger(left,right)
}
function merger(left,right){
    let res=[]
    while(left.length && right.length){
        res.push(left[0] < right[0] ? left.shift() : right.shift())
    }
    return res.concat(left,right)
}
let arr=[5,1,4,2,3]
console.log(mergeSort(arr))
```

```typescript
function slidingWIndow(nums,k){
    let queue=[],ans=[]
    for(let i=0;i<nums.length;i++){
        while(queue.length && queue[0] < i-k+1){
            queue.shift()
        }
        while(queue.length && nums[queue[queue.length-1]]<nums[i]){
            queue.pop()
        }
        queue.push(i)
        if(i >= k){
            ans.push(nums[queue[0]])
        }
    }
    return ans
}
let nums=[1,3,-1,-3,5,3,6,7], k = 3
console.log(slidingWIndow(nums,k))
```

```typescript
let binarySearchSearch=function(nums,target){
    let left=0,right=nums.length-1
    while(left < right){
        let mid=Math.floor((left+right)/2)
        if(nums[mid] === target)return mid
        else if(nums[mid] > target)right=mid-1
        else left=mid+1
    }
    return -1
}
let nums=[-1,0,3,5,9,12], target = 9
console.log(binarySearchSearch(nums,target))
```

```typescript
let countPrime=function(n){
    let isPrime=new Array(n).fill(true)
    isPrime[0]=isPrime[1]=false
    for(let i=2;i*i<n;i++){
        if(isPrime[i]){
            for(let j=i*i;j<n;j+=i){
                isPrime[j]=false
            }
        }
    }
    return isPrime.filter(prime=>prime).length
}
let n=10
console.log(countPrime(n))
```

```typescript
let searchInrotatedArray=function(nums,target){
    let left=0,right=nums.length-1
    while(left <= right){
        let mid=Math.floor((left+right)/2)
        if(nums[mid] === target)return mid
        else if(nums[mid] >= nums[left]){
            if(target <= nums[mid] && target >= nums[left]){
                right=mid-1 
            }
            else left=mid+1
        }
        else{
            if(target >= nums[mid] && target <= nums[right]){
                left=mid+1
            }
            else{
                right=mid-1
            }
        }
    }
    return -1
}
let nums = [4,5,6,7,0,1,2], target = 0
console.log(searchInrotatedArray(nums,target))
```

```typescript
function selectionSOrt(arr) {
    let count=0
    for(let i=0;i<arr.length;i++){
        let temp=i
        for(let j=i+1;j<arr.length;j++){
            if(arr[j] < arr[temp]){
                temp=j
            }
        }
        if(temp !== i){
            let ans=arr[i]
            arr[i]=arr[temp]
            arr[temp]=ans
            count++
        }
        
    }
    return count
}
let arr=[2,8,5,4]
console.log(selectionSOrt(arr))
```
