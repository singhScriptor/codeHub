```typescript
let maximumSlidingWindow=function(nums,k){
    let queue=[],ans=[]
    for(let i=0;i<nums.length;i++){
        while(queue.length && queue[0] < i-k+1){
            queue.shift()
        }
        while(queue.length && nums[queue[queue.length-1]]< nums[i]){
            queue.pop()
        }
        queue.push(i)
        if(i >= k-1){
            ans.push(nums[queue[0]])
        }
    }
    return ans
}
let  nums = [1,3,-1,-3,5,3,6,7], k = 3
console.log(maximumSlidingWindow(nums,k))
```

```typescript
let VALIDATESTACKSEQ=function(pushed,popped){
    let stack=[]
    for(let i of pushed){
        stack.push(i)
        while(stack.length && stack[stack.length-1] === popped[0]){
            stack.pop()
            popped.shift()
        }
    }
    return stack.length === 0
}
let pushed = [1,2,3,4,5], popped = [4,5,3,2,1]
console.log(VALIDATESTACKSEQ(pushed,popped))
```

```typescript
let twoSUM=function(nums,target){
    for(let i=0;i<nums.length;i++){
        for(let j=1;j<nums.length;j++){
            if((nums[i]+nums[j])=== target){
                return [i,j]
            }
        }
    }
    return [-1,-1]
}
let  nums = [2,7,11,15], target = 9
console.log(twoSUM(nums,target))
```

```typescript
let countPRIMES=function(n){
    let isPrime=new Array(n).fill(true)
    isPrime[0]=isPrime[1]=false
    for(let i=2;i*i<n;i++){
        if(isPrime[i]){
            for(let j=i*i;j<n;j+=i){
                isPrime[j]=false
            }
        }
    }
    return isPrime.filter(prime=>prime).length
}
let n=10
console.log(countPRIMES(n))
```

```typescript
let permuTATION=function(str){
    if(str.length <= 1)return [str]
    let perm=[]
    for(let i=0;i<str.length;i++){
        let char=str[i]
        let ans=str.slice(0,i)+str.slice(i+1)
        for(let j of permuTATION(ans)){
            perm.push(char+j)
        }
    }
    return perm
}
let str='abc'
console.log(permuTATION(str))
```

```typescript
let showTimes=function(n,k,movie){
    movie.sort((a,b)=>a[1]-b[1])
    let recess=new Array(k).fill(0),count=0
    for(let i=0;i<n;i++){
        let [start,end]=movie[i]
        let index=-1
        for(let j=0;j<k;j++){
            if(recess[j] <= start){
                index=j
                break
            }
        }
        if(index !== -1){
            recess[index]=end
            count++
        }
    }
    return count
}
```

```typescript
let findAverageAge=function (obj){
    let object=Object.values(obj)
    let length=object.length
    let sum=object.reduce((a,b)=>a+b,0)
    let average=sum/length
    for(let i in object){
        if(object[i] === average){
            return object[i]
        }
    }
}
let studentobj = {

'yash': 26,

'vaibhav': 24,

'rajesh' : 25,

}

console.log(findAverageAge(studentobj))
```

```typescript
let findSumLinkedList=function(head){
    let odd=0
    while(head){
        if(head.val % 2 !== 0){
            odd+=head.val
        }
        head=head.next
    }
    return odd
}
```

```typescript
let middleInLinkedList=function(head){
    let fast=head,slow=head
    while(fast && fast.next){
        fast=fast.next.next
        slow=slow.next
    }
    return slow
}
```

```typescript
delete3rdELement=function(head){
    let size=0,fast=head
    while(fast){
        size++
        fast=fast.next
    }
    let target=size-3
    let slow=head
    if(target > 0){
        for(let i=1;i<target-1;i++){
            slow=slow.next
        }
        if(slow.next){
            slow.next=slow.next.next
        }
    }
}
```

```typescript
function sortArrayClosest2Five(arr){
    return arr.sort((a,b)=>{
        let A=Math.abs(a-5)
        let B=Math.abs(b-5)
        return A-B 
    })
}
let arr=[1,3,5,7,9]
console.log(sortArrayClosest2Five(arr))
```

```typescript
function add2Matrices(ar1,ar2){
    let arr=[]
    for(let i=0;i<ar1.length;i++){
        arr[i]=[]
        for(let j=0;j<ar1[0].length;j++){
            arr[i][j]=ar1[i][j]+ar2[i][j]
        }
    }
    return arr
}
let ar1=[[1,2,5],[3,5,5]],ar2=[[1,5,7],[2,1,3]]
console.log(add2Matrices(ar1,ar2))
```

```typescript
let minimumSwap=function(arr){
    let count=0
    for(let i=0;i<arr.length;i++){
        let temp=i
        for(let j=i+1;j<arr.length;j++){
            if(arr[j] < arr[temp]){
                temp=j
            }
        }
        if(temp !== i){
            let ans=arr[i]
            arr[i]=arr[temp]
            arr[temp]=ans
            count++
        }
        
    }
    return count
}
let arr=[10,19,6,3,5]
console.log(minimumSwap(arr))
```
