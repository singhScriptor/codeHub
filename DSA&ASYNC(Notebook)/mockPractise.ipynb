```typescript
function validParentheses(s){
    let dict={']':'[','}':'{',')':'('}
    let stack=[]
    for(let i of s){
        if(i in dict){
            if(stack.length === 0 || stack.pop() !== dict[i])return false 
        }
        else stack.push(i)
    }
    return stack.length === 0
}
let s="{}"
console.log(validParentheses(s))
```

```typescript
let binarySearch=function(nums,target){
    let left=0,right=nums.length-1
    while(left <= right){
        let mid=Math.floor((left+right)/2)
        if(nums[mid] === target)return mid 
        else if(nums[mid] > target)right=mid-1 
        else left=mid+1
    }
    return -1
}
let nums=[-1,0,3,5,9,12],target=9
console.log(binarySearch(nums,target))
```

```typescript
let trappingRainWater=function(heights){
    let left=0,right=heights.length-1,leftMax=heights[left],rightMax=heights[right],
    ans=0
    while(left < right){
        if(leftMax < rightMax){
            left++
            leftMax=Math.max(leftMax,heights[left])
            ans+=leftMax-heights[left]
        }
        else{
            right--
            rightMax=Math.max(rightMax,heights[right])
            ans+=rightMax-heights[right]
        }
    }
    return ans
}
let height = [0,1,0,2,1,0,1,3,2,1,2,1]
console.log(trappingRainWater(height))
```

```typescript
let mergeSort=function(arr){
    if(arr.length <= 1) return arr
    let mid=Math.floor(arr.length/2)
    let left=mergeSort(arr.slice(0,mid))
    let right=mergeSort(arr.slice(mid))
    return merger(right,left)

}
function merger(right,left){
    let ans=[]
    while(left.length && right.length){
        ans.push(left[0] < right [0] ? left.shift() : right.shift())
    }
    return ans.concat(left,right)
}
let arr=[4,1,3,5,2]
console.log(mergeSort(arr))
```

```typescript
function greatesNumberCandies(candies,extra){
    let maxi=-Infinity
    for(let i=0; i< candies.length;i++){
        if(candies[i] > maxi){
            maxi=candies[i]
        }
    }
    let arr=[]
    for(let i=0;i<candies.length;i++){
        if(candies[i] + extra >= maxi){
            arr.push(true)
        }
        else{
            arr.push(false)
        }
    }
    return arr
}
let candies = [2,3,5,1,3], extraCandies = 3
console.log(greatesNumberCandies(candies,extraCandies))
```

```typescript
function concatenationArray(nums){
    let arr=[]
    for(let i=0;i<nums.length;i++){
        arr.push(nums[i])
    }
    for(let i=0;i<nums.length;i++){
        arr.push(nums[i])
    }
    return arr
}
let nums=[1,2,1]
console.log(concatenationArray(nums))
```

```typescript
let shuffleArray=function(nums,n){
    let arr=[]
    for(let i=0;i<n;i++){
        arr.push(nums[i])
        arr.push(nums[i+n])
    }
    return arr
}
let nums = [2,5,1,3,4,7], n = 3
console.log(shuffleArray(nums,n))
```

```typescript
function sortAnArray(nums){
    if(nums.length <= 1)return nums
    let mid=Math.floor(nums.length/2)
    let left=sortAnArray(nums.slice(0,mid))
    let right=sortAnArray(nums.slice(mid))
    return sorted(left,right)
}
function sorted(left,right){
    let ans=[]
    while(left.length && right.length){
        ans.push(left[0] < right[0] ? left.shift() : right.shift())
    }
    return ans.concat(left,right)
}
let nums=[5,2,3,1]
console.log(sortAnArray(nums))
```

```typescript
let validateStackSeq=function(pushed,popped){
    let stack=[]
    for(let i of pushed){
        stack.push(i)
        while(stack.length && stack[stack.length-1] === popped[0]){
            stack.pop()
            popped.shift()
        }
    }
    return stack.length === 0
}
```

```typescript
let countPrime=function(n){
    let isPrime=new Array(n).fill(true)
    isPrime[0]=isPrime[1]=false
    for(let i=2;i*i<n;i++){
        if(isPrime[i]){
            for(let j=i*i;j<n;j+=i){
                isPrime[j]=false
            }
        }
    }
    return isPrime.filter(prime=>prime).length
}
let n=10
console.log(countPrime(n))
```

```typescript
let delete3rdElement=function(head){
    let fast=head,size=0
    while(fast){
        size++
        fast=fast.next
    }
    let target=size-3
    let slow=head
    if(target > 0){
        for(let i=1;i<target;i++){
            slow=slow.next
        }
        if(slow.next){
            slow.next=slow.next.next
        }
    }
}
```

```typescript
let nonRepeatingChar=function(A){
    let freq=new Array(26).fill(0)
    let ans='',stack=[]
    for(let i of A){
        let index=i.charCodeAt(0)-'a'.charCodeAt(0)
        freq[index]++
        stack.push(i)
        while(stack.length && freq[stack[0].charCodeAt(0)-'a'.charCodeAt(0)]>1){
            stack.shift()
        }
        if(stack.length){
            ans+=stack[0]
        }
        else{
            ans+='#'
        }
    }
    return ans
}
```
