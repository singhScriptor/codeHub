```typescript
let decodeString=function(s){
    let stack=[],str='',num=0
    for(let i of s){
        if(!isNaN(i)){
            num=num*10+parseInt(i)
        }
        else if(i === '['){
            stack.push(str)
            stack.push(num)
            str=''
            num=0
        }
        else if(i === ']'){
            let prevNum=stack.pop()
            let prevStr=stack.pop()
            str=prevStr+str.repeat(prevNum)
        }
        else{
            str+=i
        }
    }
    return str
}
let s="3[a]2[bc]"
console.log(decodeString(s))
```

```typescript
let binarySearch=function(nums,target){
    let left=0,right=nums.length-1
    while(left < right){
        let mid=Math.floor((left+right)/2)
        if(nums[mid] === target){
            return mid
        }
        else if(nums[mid] > target)right=mid-1 
        else left=mid+1
    }
    return -1
}
let nums = [-1,0,3,5,9,12], target = 9
console.log(binarySearch(nums,target))
```

```typescript
let maxSlidingWindow=function(nums,k){
    let queue=[],ans=[]
    for(let i=0;i<nums.length;i++){
        while(queue.length && queue[0] < i-k+1){
            queue.shift()
        }
        while(queue.length && nums[queue[queue.length-1]] < nums[i]){
            queue.pop()
        }
        queue.push(i)
        if(i >= k-1){
            ans.push(nums[queue[0]])
        }
    }
    return ans
}
let  nums = [1,3,-1,-3,5,3,6,7], k = 3
console.log(maxSlidingWindow(nums,k))
```

```typescript
let largestRectHisto=function(heights){
    heights.push(0)
    let stack=[-1],maxi=0
    for(let i=0;i<heights.length;i++){
        while(heights[i] < heights[stack[stack.length-1]]){
            let left=heights[stack.pop()]
            let right=i-stack[stack.length-1]-1
            maxi=Math.max(maxi,(left*right))
        }
        stack.push(i)
    }
    return maxi
}
let heights = [2,1,5,6,2,3]
console.log(largestRectHisto(heights))
```

```typescript
let trappingRainWater=function(height){
    let left=0,right=height.length-1,
    leftMax=height[left],rightMax=height[right],
    ans=0
    while(left < right){
        if(leftMax < rightMax){
            left++
            leftMax=Math.max(leftMax,height[left])
            ans+=leftMax-height[left]
        }
        else{
            right--
            rightMax=Math.max(rightMax,height[right])
            ans+=rightMax-height[right]
        }
    }
    return ans
}
let  height = [0,1,0,2,1,0,1,3,2,1,2,1]
console.log(trappingRainWater(height))
```

```typescript
let decodeString=function(s){
    let stack=[],num=0,str=''
    for(let i of s){
        if(!isNaN(i)){
            num=num*10+parseInt(i)
        }
        else if(i === '['){
            stack.push(str)
            stack.push(num)
            str=''
            num=0
        }
        else if(i === ']'){
            let prevNum=stack.pop()
            let prevStr=stack.pop()
            str=prevStr+str.repeat(prevNum)
        }
        else{
            str+=i
        }
    }
    return str
}
let s = "3[a]2[bc]"
console.log(decodeString(s))
```

```typescript
let validateStackSeq=function(pushed,popped){
    let stack=[]
    for(let i of pushed){
        stack.push(i)
        while(stack.length && stack[stack.length-1] === popped[0]){
            stack.pop()
            popped.shift()
        }

    }
    return stack.length === 0
}
let pushed = [1,2,3,4,5], popped = [4,5,3,2,1]
console.log(validateStackSeq(pushed,popped))
```

```typescript
let richerCustomer=function(accounts){
    let maxi=0
    for(let i=0;i<accounts.length;i++){
        let total=accounts[i].reduce((a,b)=>a+b,0)
        maxi=Math.max(maxi,total)
    }
    return maxi
}
let accounts = [[1,2,3],[3,2,1]]
console.log(richerCustomer(accounts))
```

```typescript
let findLucky=function(arr){
    let set=new Set(arr),maxi=-1
    for(let i of set){
        let index=arr.filter(sh=> sh === i).length
        if(index === i){
            maxi=Math.max(maxi,i)
        }
    }
    return maxi
}
let arr=[2,2,3,4]
console.log(findLucky(arr))
```

```typescript
let twoSum=function(nums,target){
    for(let i=0;i<nums.length;i++){
        for(let j=i+1;j<nums.length;j++){
            if((nums[i]+nums[j])=== target){
                return [i,j]
            }
        }
    }
    return [-1,-1]
}
let nums=[2,7,11,15],target=9
console.log(twoSum(nums,target))
```

```typescript
let binarySearch=function(nums,target){
    let left=0,right=nums.length-1
    while(left <= right){
        let mid=Math.floor((left+right)/2)
        if(nums[mid] === target)return mid
        else if(nums[mid] > target)right=mid-1
        else left=mid+1 
    }
    return -1
}
let nums = [-1,0,3,5,9,12], target = 9
console.log(binarySearch(nums,target))
```

```typescript
let rotateArray=function(nums,k){
    k=k%nums.length
    let rotated=nums.slice(-k).concat(nums.slice(0,-k))
    for(let i=0;i<nums.length;i++){
        nums[i]=rotated[i]
    }
    return nums
}
let nums = [1,2,3,4,5,6,7], k = 3
console.log(rotateArray(nums,k))
```

```typescript
let transposeMatrix=function(matrix){
    let ans=[]
    for(let i=0;i<matrix[0].length;i++){
        ans[i]=[]
        for(let j=0;j<matrix.length;j++){
            ans[i].push(matrix[j][i])
        }
    }
    return ans
}
let matrix = [[1,2,3],[4,5,6],[7,8,9]]
console.log(transposeMatrix(matrix))
```

```typescript
let sortByOnesBit=function(arr){
    return arr.sort((a,b)=>{
        let A=a.toString(2).split('').filter(sh=>sh==='1').length
        let B=b.toString(2).split('').filter(sh=>sh==='1').length
        return A-B || a-b
    })
}
let  arr = [0,1,2,3,4,5,6,7,8]
console.log(sortByOnesBit(arr))
```
