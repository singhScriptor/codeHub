```typescript
let maxSlidingWindow=function(nums,k){
    let queue=[],ans=[]
    for(let i=0;i<nums.length;i++){
        while(queue.length && queue[0] < i-k+1){
            queue.shift()
        }
        while(queue.length && nums[queue[queue.length-1]]<nums[i]){
            queue.pop()
        }
        queue.push(i)
        if(i >= k-1){
            ans.push(nums[queue[0]])
        }
    }
    return ans
}
let nums = [1,3,-1,-3,5,3,6,7], k = 3
console.log(maxSlidingWindow(nums,k))
```

```typescript
let validParentheses=function(s){
    let dict={']':'[','}':'{',')':'('},
    stack=[]
    for(let i of s){
        if(i in dict){
            if(stack.length === 0 || stack.pop() !== dict[i])return false
        }
        else stack.push(i)
    }
    return stack.length === 0
}
let s="{}"
console.log(validParentheses(s))
```

```typescript
let largestRectangleHisto=function(height){
    height.push(0)
    let maxi=0,stack=[-1]
    for(let i=0;i<height.length;i++){
        while(height[i] < height[stack[stack.length-1]]){
            let left=height[stack.pop()],
            right=i-stack[stack.length-1]-1
            maxi=Math.max(maxi,(left*right))
        }
        stack.push(i)
    }
    return maxi
}
let heights = [2,1,5,6,2,3]
console.log(largestRectangleHisto(heights))
```

```typescript
let trappingRainWater=function(heights){
    let left=0,right=heights.length-1,
    leftMax=heights[left],rightMax=heights[right],
    ans=0
    while(left < right){
        if(leftMax < rightMax){
            left++
            leftMax=Math.max(leftMax,heights[left])
            ans+=leftMax-heights[left]
        }
        else{
            right--
            rightMax=Math.max(rightMax,heights[right])
            ans+=rightMax-heights[right]
        }
    }
    return ans
}
let height = [4,2,0,3,2,5]
console.log(trappingRainWater(height))
```

```typescript
let nextGreaterElement=function(nums){
    let arr=new Array(nums.length)
    for(let i=0;i<nums.length;i++){
        arr[i]=-1
        for(let j=1;j<nums.length;j++){
            let index=(i+j)%nums.length
            if(nums[index] > nums[i]){
                arr[i]=nums[index]
            }
        }
    }
    return arr
}
let nums=[1,2,1]
console.log(nextGreaterElement(nums))
```

```typescript
let decodeString=function(s){
    let stack=[],num=0,str=''
    for(let i of s){
        if(!isNaN(i)){
            num=num*10+parseInt(i)
        }
        else if(i === '['){
            stack.push(str)
            stack.push(num)
            str=''
            num=0
        }
        else if(i === ']'){
            let prevNum=stack.pop()
            let prevStr=stack.pop()
            str=prevStr+str.repeat(prevNum)
        }
        else{
            str+=i
        }
    }
    return str
}
let  s = "3[a]2[bc]"
console.log(decodeString(s))
```

```typescript
let adjacentDuplicate=function(s){
    let stack=[]
    for(let i of s){
        if(stack.length && stack[stack.length-1]===i){
            stack.pop()
        }
        else stack.push(i)
    }
    return stack.join('')
}
let s="abbaca"
console.log(adjacentDuplicate(s))
```

```typescript
let validateStackSeq=function(pushed,popped){
    let stack=[]
    for(let i of pushed){
        stack.push(i)
        while(stack.length && stack[stack.length-1] === popped[0]){
            stack.pop()
            popped.shift()
        }
    }
    return stack.length === 0
}
let pushed = [1,2,3,4,5], popped = [4,5,3,2,1]
console.log(validateStackSeq(pushed,popped))
```

```typescript
let richerCustomerWealth=function(accounts){
    let maxi=-1
    for(let i of accounts){
        let total=i.reduce((a,b)=>a+b,0)
        maxi=Math.max(maxi,total)
    }
    return maxi
}
let accounts= [[1,2,3],[3,2,1]]
console.log(richerCustomerWealth(accounts))
```

```typescript
let countPrime=function(n){
    let isPrime=new Array(n).fill(true)
    isPrime[0]=isPrime[1]=false
    for(let i=2;i*i<n;i++){
        if(isPrime[i]){
            for(let j=i*i;j<n;j+=i){
                isPrime[j]=false
            }
        }
    }
    return isPrime.filter(prime=>prime).length
}
let n=10
console.log(countPrime(n))
```

```typescript
let nNumberPrimes=function(n){
    let x=2,arr=[]
    while(arr.length < n){
        let flag=true
        for(let i=2;i<x;i++){
            if(x%i===0){
                flag=false
                break
            }
        }
        if(flag){
            arr.push(x)
        }
        x++
    }
    return arr
}
let n=5
console.log(nNumberPrimes(n))
```

```typescript
let luckyInteger=function(arr){
    let set=new Set(arr),maxi=-1
    for(let i of set){
        let index=arr.filter(sh=>sh === i).length
        if(i === index){
            maxi=Math.max(maxi,i)
        }
    }
    return maxi
}
let arr=[2,2,3,4]
console.log(luckyInteger(arr))
```

```typescript
let twoSum=function(nums,target){
    for(let i=0;i<nums.length;i++){
        for(let j=1;j<nums.length;j++){
            if(nums[i]+nums[j]===target){
                return [i,j]
            }
        }
    }
    return [-1,-1]
}
let nums = [2,7,11,15], target = 9
console.log(twoSum(nums,target))
```

```typescript
let binarySearch=function(nums,target){
    let left=0,right=nums.length-1
    while(left < right){
        let mid=Math.floor((left+right)/2)
        if(nums[mid]===target)return mid
        else if(nums[mid]> target)right=mid-1
        else left=mid+1
    }
    return -1
}
let nums = [-1,0,3,5,9,12], target = 9
console.log(binarySearch(nums,target))
```

```typescript
let transposeMatrix=function(matrix){
    let column=matrix[0].length,row=matrix.length,
    ans=[]
    for(let i=0;i< column;i++){
        ans[i]=[]
        for(let j=0;j<row;j++){
            ans[i].push(matrix[j][i])
        }
    }
    return ans
}
let matrix = [[1,2,3],[4,5,6],[7,8,9]]
console.log(transposeMatrix(matrix))
```

```typescript
let insertionSort=function(arr){
    for(let i=0;i<arr.length;i++){
        let temp=arr[i],j=i-1 
        while(j>=0 && arr[j] > temp){
            arr[j+1]=arr[j]
            j--
        }
        arr[j+1]=temp
    }
    return arr
}
let arr=[5,1,4,2,3]
console.log(insertionSort(arr))
```

```typescript
let maxConsecutiveOnes=function(nums){
    let maxi=-1,count=0
    for(let i=0;i<nums.length;i++){
        if(nums[i] === 1){
            count++
            maxi=Math.max(maxi,count)
        }
        else count=0
    }
    return maxi
}
let  nums = [1,1,0,1,1,1]
console.log(maxConsecutiveOnes(nums))
```
