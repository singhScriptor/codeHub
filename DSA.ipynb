```typescript
let slidingWindow=function(nums,k){
    let queue=[],ans=[]
    for(let i=0;i<nums.length;i++){
        while(queue.length && queue[0] < i-k+1){
            queue.shift()
        }
        while(queue.length && nums[queue[queue.length-1]]<nums[i]){
            queue.pop()
        }
        queue.push(i)
        if(i>=k-1){
            ans.push(nums[queue[0]])
        }
    }
    return ans
}
let nums=[1,3,-1,-3,5,3,6,7], k = 3
console.log(slidingWindow(nums,k))
```

```typescript
let trappingRainWater=function(heights){
    let left=0,right=heights.length-1,
    leftMax=heights[left],rightMax=heights[right],ans=0
    while(left < right){
        if(leftMax < rightMax){
            left++
            leftMax=Math.max(leftMax,heights[left])
            ans+=leftMax-heights[left]
        }
        else{
            right--
            rightMax=Math.max(rightMax,heights[right])
            ans+=rightMax-heights[right]
        }
    }
    return ans
}
let height = [0,1,0,2,1,0,1,3,2,1,2,1]
console.log(trappingRainWater(height))
```

```typescript
let decodeString=function(s){
    let str='',num=0,stack=[]
    for(let i of s){
        if(!isNaN(i)){
            num=num*10+parseInt(i)
        }
        else if(i === '['){
            stack.push(str)
            stack.push(num)
            str=''
            num=0
        }
        else if(i === ']'){
            let prevNum=stack.pop()
            let prevStr=stack.pop()
            str=prevStr+str.repeat(prevNum)
        }
        else{
            str+=i
        }
    }
    return str
}
let s="3[a]2[bc]"
console.log(decodeString(s))
```

```typescript
let findMaxWOrd=function(sentence){
    let maxi=0
    for(let i of sentence){
        let total=i.split(' ')
        maxi=Math.max(maxi,total.length)
    }
    return maxi
}
let sentences = ["alice and bob love leetcode", "i think so too", "this is great thanks very much"]
console.log(findMaxWOrd(sentences))
```
